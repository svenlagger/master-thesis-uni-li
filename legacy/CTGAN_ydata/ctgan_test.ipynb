```python
from ydata_synthetic.synthesizers.regular import RegularSynthesizer
from ydata_synthetic.synthesizers import ModelParameters, TrainParameters
import pandas as pd
```

```python
data = pd.read_csv('credit_score_syn_interest_rate.csv')
```

```python
data['Changed_Credit_Limit'] = pd.to_numeric(data['Changed_Credit_Limit'], errors='coerce')
data = data.dropna(subset=['Changed_Credit_Limit'])

num_cols = list(data.select_dtypes(include='number').columns)
cat_cols = ['Month','Occupation', 'Type_of_Loan', 'Credit_Mix', 'Payment_of_Min_Amount', 'Payment_Behaviour']

# data = data.drop(['Unnamed: 0', 'ID', 'Customer_ID', 'Name', 'SSN'], axis=1)
```

```python
# Defining the training parameters
batch_size = 500
epochs = 500+1
learning_rate = 2e-4
beta_1 = 0.5
beta_2 = 0.9

ctgan_args = ModelParameters(batch_size=batch_size,
                             lr=learning_rate,
                             betas=(beta_1, beta_2))

train_args = TrainParameters(epochs=epochs)
synth = RegularSynthesizer(modelname='ctgan', model_parameters=ctgan_args)
print('Fitting begins...')
synth.fit(data=data, train_arguments=train_args, num_cols=num_cols, cat_cols=cat_cols)
print('Fitting completed!')

# synth.save('adult_ctgan_model.pkl')

#########################################################
#    Loading and sampling from a trained synthesizer    #
#########################################################
# synth = RegularSynthesizer.load('adult_ctgan_model.pkl')
synth_data = synth.sample(1000)
print(synth_data)
```

```python
synth_data_large = synth.sample(10000)
```
